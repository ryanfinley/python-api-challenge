```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st

import seaborn as sns
from datetime import datetime

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#create a base and indicate imperial units for Fahrenheit
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
```

```python
# build query URL
city2 = "Atlanta"
query_url = url + "appid=" + weather_api_key + "&q=" + city2 + "&units=" + units
testResponse2 = requests.get(query_url).json()
testResponse2
```

```python
#set counters
recordCount = 1
setCount = 1
columns = ["City", "Lat", "Lng", "Max Temp (F)", "Humidity", "Wind Speed", "Cloudiness", "Country", "Date"]
cityWeather =pd.DataFrame(columns = columns)

# set iteration
for x, city in enumerate(cities):
    if (x % 50 ==0 and x >= 50):
        setCount += 1
        recordCount = 0
    # Include a print log of each city as it'sbeing processed (with the city number and city name).   
    print(f"Processing Record {recordCount} of Set {setCount} | {city}")
    recordCount +=1
    
    query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=" + units
    
    try:
        weatherResponse = requests.get(query_url).json()
        lon = weatherResponse['coord']['lon']
        lat = weatherResponse['coord']['lat']
        city_max_temp = weatherResponse['main']['temp_max']
        city_humidity = weatherResponse['main']['humidity']
        city_wind_speed = weatherResponse['wind']['speed']
        city_clouds = weatherResponse['clouds']['all']
        country = weatherResponse['sys']['country']
        date = weatherResponse['dt']
        
        #append city into cityWeather
        cityWeather = cityWeather.append({
            "City":city,
            "Lat":lat,
            "Lng":lon,
            "Max Temp (F)":city_max_temp,
            "Humidity":city_humidity,
            "Wind Speed":city_wind_speed,
            "Cloudiness":city_clouds,
            "Country":country,
            "Date":date
        }, ignore_index=True)
        
    except Exception as e:
        print(e)
        pass

```

## Linear Regression

```python
# Define northern and southern dataframes
northernDF = cityWeather.loc[pd.to_numeric(cityWeather["Lat"]).astype(float)>0,:]
#northernDF.head()
southernDF = cityWeather.loc[pd.to_numeric(cityWeather["Lat"]).astype(float)<0,:]
#southernDF.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#establish x and y values and find St. Pearson Correlation Coefficient for Mouse Weight and Tumor Volume Avg
correlation = round(st.pearsonr(northernDF['Lat'], northernDF['Max Temp (F)'])[0],2)

#print St. Pearson Correlation Coefficient
print(f"The correlation between both factors is {correlation:.2f}.")

model = st.linregress(northernDF['Lat'], northernDF['Max Temp (F)'])
print(model)

#linear regression line (y =mx+b)
lineEquation = northernDF['Lat']*model[0] + model[1]

# scatter plot of the data
plt.scatter(northernDF['Lat'], northernDF['Max Temp (F)'])
plt.title("Max Temp vs Latitdude")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")

#draw the line with the linear regression line on the plot
plt.plot(northernDF['Lat'], lineEquation, color="red")

plt.savefig("N_hemi_lat_v_max_temp.png")
plt.show()
```

```python
The max temp vs latitude plot indicates that the max temperature decreases the further a location is from the equator in the northern hemisphere.
```
